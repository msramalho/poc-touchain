```python
from lxml import etree
import lxml.html
import html as html_decoder
from bs4 import BeautifulSoup
from unidecode import unidecode
import re
from slugify import slugify
```

```python
with open("whoiswho.html", encoding="utf-8") as inf:
    doc = lxml.html.fromstring(inf.read())
```

```python
def clean_list(l):
    return list(filter(lambda x:len(x)>0, l))
```

```python
def clean_span(s):
    return " ".join(clean_list(re.split('\s{2,}', s[0].text_content()))).strip()
```

```python
nodes, edges = [], []
def parse(el, depth, parent):
    classes = el.get('class')
    if 'tree-container' in classes: # ul with <li> inside
        for tn in el.xpath("./li[@class='tree-node']"):
            parse(tn, depth+1, parent)
            
    elif 'tree-node' in classes: # li with
#         print(el.xpath(".//div[@class='tree-node-content']//text()")) // ardeu
        
        maybe_entidade = el.xpath("./div/span[@class='tree-label']//text()")
        if len(maybe_entidade) == 1: 
            entidade = [clean_span(el.xpath("./div/span[@class='tree-label']")).strip(), "ORG"]
        else: 
            entidade = []
        
            name = el.cssselect("span.wiw-sublevel-person-name")
            if len(name): entidade.append(clean_span(name))

            position = el.cssselect("span.wiw-sublevel-person-field")
            if len(position): entidade.append(clean_span(position))

            phone = el.cssselect("span.wiw-sublevel-person-phone")
            if len(phone): entidade.append(clean_span(phone))
            
            entidade.append("PER")

        nodes.append([slugify(entidade[0])] + [entidade])
        edges.append([slugify(parent[0]), slugify(entidade[0])])
        text = ",".join(entidade).strip() if len(entidade) else None
        if text: print(" " * depth, text)
        for new_el in el.xpath("./ul[@class='tree-container']"):
            parse(new_el, depth+1, entidade)
            
base = doc.cssselect("ul.tree-container")[0]
parse(base, 0, ["root"])
```
